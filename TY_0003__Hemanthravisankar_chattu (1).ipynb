# import dataset and libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/supply_chain_data.csv')
```

```python
df
```

```python
df.info()
```

# line plot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.plot(df['Product type'], df['Revenue generated'], marker='o', linestyle='-')
plt.title('Revenue Generated for Each Product Type')
plt.xlabel('Product Type')
plt.ylabel('Revenue Generated')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# scatter plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Number of products sold'], df['Revenue generated'])
plt.title('Scatter Plot of Number of Products Sold vs. Revenue Generated')
plt.xlabel('Number of Products Sold')
plt.ylabel('Revenue Generated')
plt.show()
```

# histogram plot

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Price'], bins=20, edgecolor='black')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

# box plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.boxplot(x='Transportation modes', y='Defect rates', data=df)
plt.title('Box Plot of Defect Rates by Transportation Modes')
plt.xlabel('Transportation Modes')
plt.ylabel('Defect Rates')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# pie chart

```python
shipping_carriers_counts = df['Shipping carriers'].value_counts()
plt.pie(shipping_carriers_counts, labels=shipping_carriers_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Shipping Carriers Distribution')
plt.show()
```

# bar chart


```python
customer_demographics_counts = df['Customer demographics'].value_counts()
plt.bar(customer_demographics_counts.index, customer_demographics_counts)
plt.title('Customer Demographics Distribution')
plt.xlabel('Customer Demographics')
plt.ylabel('Count')
plt.show()
```
